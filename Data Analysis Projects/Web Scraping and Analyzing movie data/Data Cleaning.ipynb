```python
def find_date(x):
    global y
    try:
        if re.search('(\d\d\d\d-\d\d-\d\d)', x) != None:
            y = re.findall('(\d\d\d\d-\d\d-\d\d)', x)[0]

        elif re.search('[a-zA-Z]{1,10} \d{1,2}, \d\d\d\d',x) != None:
            y = re.findall('[a-zA-Z]{1,10} \d{1,2}, \d\d\d\d',x)[0]

        elif re.search('(\d\d\d\d)', x) != None:
            y = re.findall('(\d\d\d\d)',x)[0]
    except TypeError:
        y = np.NaN

    
    return y
```

```python
df['release_date'] = pd.to_datetime(df['Release date'].apply(find_date))
```

```python
countries = []
for i in df['Country'].unique():
    try:
        countries.append(re.findall('[A-Z]+[a-z]+',i))
    except:
        countries.append(np.NaN)
```

```python
for k in range(2): # repeating the loop twice because because two countries with United in it are present 
    # i.e (UnitedStates and UnitedKingdom)
    for i in countries:
        try:
            if 'United' in i:
                in_ = i.index('United')
                z=(' '.join(i[in_:in_+2]))
                
                for n in range(2):
                    i.pop(in_)  # first remove united, now at the index of united in_+1 value will come remove that as well

                i.append(z)
        except:
            continue
```

Creating a Function for country

```python
def convert_country(val):
    global i
    try:
            i=(re.findall('[A-Z]+[a-z]+',val))
    except:
            i=(np.NaN)


    for k in range(2):
        try:
            if 'United' in i:
                in_ = i.index('United')
                z=(' '.join(i[in_:in_+2]))

                for n in range(2):
                    i.pop(in_)
                i.append(z)
        except:
            continue
    return i
```
